```python
import pandas as pd
import numpy as np
```

```python
rank2016 = pd.read_excel("/Ranking_2016.xlsx", sheet_name= None)
rank2017 = pd.read_excel("/Ranking_2017.xlsx", sheet_name= None)
rank2018 = pd.read_excel("/Ranking_2018.xlsx", sheet_name= None)
rank2019 = pd.read_excel("/Ranking_2019.xlsx", sheet_name= None)
rank2020 = pd.read_excel("/Ranking_2020.xlsx", sheet_name= None)
rank2021 = pd.read_excel("/Ranking_2021.xlsx", sheet_name= None)
rank2022 = pd.read_excel("/Ranking_2022.xlsx", sheet_name= None)
rank2023 = pd.read_excel("/Ranking_2023.xlsx", sheet_name= None)
```

```python
ranks = [rank2016,rank2017,rank2018,rank2019,rank2020,rank2021,rank2022,rank2023]
```

```python
for rank in ranks:
  key = rank.keys()
  if ('Índice' in key):
    rank.pop('Índice')
  if ('Descrições' in key):
    rank.pop('Descrições')
  if ('Geral' in key):
    rank.pop('Geral')
  if ('Auxiliar' in key):
    rank.pop('Auxiliar')
  print(key)
```

```python
tables = []
dicio = {}
for rank in ranks:
  key = rank.keys()
  for i in key:
    des_table = pd.DataFrame(rank[i].iloc[:31, 5:])
    label = des_table.iloc[2].values
    label = label[1:]
    label = np.insert(label, 0,'Estado')
    des_table.columns = label
    org_table = des_table.drop([2,3])
    coluna = org_table.iloc[1][ org_table.iloc[1] == org_table.iloc[1].agg('max') ].index[0]

    coluna = org_table[coluna]
    coluna = coluna.to_frame()
    if coluna.columns[0]>=2015:
      if i not in dicio:
        dicio[i] = coluna
      if dicio.get(i).columns[0] != coluna.columns[0]:
        dicio.get(i).insert(0 ,coluna.columns[0], coluna)

with pd.ExcelWriter('dados.xlsx') as writer:
  for i in key:
    dicio[i].to_excel(writer, sheet_name=i, index=False)




```
